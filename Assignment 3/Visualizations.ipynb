```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Function to calculate efficiency based on probability and number of stations
def calculate_efficiency(p, n_stations):
    # Efficiency increases until p = 1/n_stations and then decreases
    return np.where(p <= (1 / n_stations),
                    (n_stations * p) / (1 + (n_stations - 1) * p),  # Increase phase
                    (n_stations / (1 + (n_stations - 1))) * (1 / (1 + (p - (1 / n_stations)) * (n_stations - 1))))  # Decrease phase

# Define the probabilities and the number of stations
probabilities = np.linspace(0.01, 1.0, 100)  # from 0.01 to 1.0
stations = [2, 4, 5]  # Number of stations to analyze

# Create the figure with increased width
fig = plt.figure(figsize=(16, 8))  # Increase width
ax = fig.add_subplot(111, projection='3d')

# Prepare to store efficiency values for each number of stations
for n in stations:
    efficiency = calculate_efficiency(probabilities, n)
    
    # Z values for the current number of stations
    z = np.full(efficiency.shape, n)  # Same n value across the length of efficiency
    
    # Scatter plot for the current station count
    scatter = ax.scatter(probabilities, efficiency, z, s=50, label=f'Stations: {n}', alpha=0.7)
    
    # Connect the points with a line
    ax.plot(probabilities, efficiency, z, linewidth=2)

# Labels and title
ax.set_title('Efficiency for p-Persistent CSMA/CD in 3D View', fontsize=18, color='darkblue')
ax.set_xlabel('Probability (p)', fontsize=14, color='darkgreen')
ax.set_ylabel('Efficiency', fontsize=14, color='darkgreen')
ax.set_zlabel('Number of Stations', fontsize=14, color='darkgreen')

# Adjust Z-axis label padding for better visibility
ax.zaxis.labelpad = 15  # Increase padding for Z-axis label

# Customize the viewing angle
ax.view_init(elev=30, azim=210)  # Adjust angles for optimal viewing

# Add a legend
ax.legend()

# Show grid
ax.grid(True)

# Enhance the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate efficiency based on probability and number of stations
def calculate_efficiency(p, n_stations):
    efficiency = (n_stations * p) / (1 + (n_stations - 1) * p)
    return efficiency

# Define the probabilities and the number of stations
probabilities = np.arange(0.01, 1.01, 0.01)  # from 0.01 to 1
stations = [2, 4, 5]

# Create a plot
plt.figure(figsize=(12, 8))

# Calculate efficiency for each number of stations and plot
for n in stations:
    efficiency = calculate_efficiency(probabilities, n)
    plt.plot(probabilities, efficiency, label=f'Stations: {n}', marker='o', linestyle='-')

# Customize the plot
plt.title('Efficiency of p-Persistent CSMA/CD for Different Numbers of Stations', fontsize=16)
plt.xlabel('Probability (p)', fontsize=14)
plt.ylabel('Efficiency', fontsize=14)
plt.xticks(np.arange(0, 1.1, 0.1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True)
plt.axvline(x=1/2, color='blue', linestyle='--', label='Max Efficiency at p=1/2')  # For 2 stations
plt.axvline(x=1/4, color='orange', linestyle='--', label='Max Efficiency at p=1/4')  # For 4 stations
plt.axvline(x=1/5, color='green', linestyle='--', label='Max Efficiency at p=1/5')  # For 5 stations

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
